```python
import pandas as pd
import datetime
```

```python
df=pd.read_csv('./2021.5~2021.6.csv')
```

```python
def make_table(df):
    df=df.drop(range(1,len(df),2))
    df=df.drop('옵션',axis=1)
    df['판매일']=pd.to_datetime(df['판매일'])
    df['기간']=df['기간'].str.replace('일','')
    df['기간']=df['기간'].str.replace('없음','0')
    df['기간']=df['기간'].fillna('0')
    df['기간']=df['기간'].astype(int)
    df['연식']=df['연식'].str.split('.')
    df['주행거리']=df['주행거리'].str.replace('Km','').str.replace(',','')
    df['주행거리'].fillna('0',inplace=True)
    df['주행거리']=df['주행거리'].astype(int)
    df['가격']=df['가격'].str.replace(',','')
    df['가격'].fillna(0,inplace=True)
    df['가격']=df['가격'].astype(int)
    return df
```

```python
a=make_table(df)
a
```
