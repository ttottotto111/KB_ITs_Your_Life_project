```python
import os
import math
import numpy as np
import tensorflow as tf
import pandas as pd
```

```python
df=pd.read_csv('./중고차데이터최종.csv',encoding='cp949')
```

```python
df=df.drop(['Unnamed: 0','index'],axis=1)
```

```python
candidates=df[['제조회사','모델명','세부모델','연료','차종','생산지']]
```

```python
sentences=[]
for i in range(len(candidates)):
    sentence=[]
    for j in range(6):
        sentence.append(candidates.iloc[i][j])
    sentences.append(sentence)
```

```python
sentences
```

```python
from gensim.models import Word2Vec
```

```python
model=Word2Vec(sentences, vector_size=100, window=4, min_count=3, workers=3,sg=1)
```

```python
model.wv.get_vector('S클래스',norm=True)
```

```python
model.wv.most_similar('S클래스 6세대',topn=100)
```
