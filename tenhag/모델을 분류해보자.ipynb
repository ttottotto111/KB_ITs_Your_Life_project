```python
import pandas as pd
```

```python
df = pd.read_csv("./중고차데이터최종v7.csv", encoding="cp949")
```

```python
df
```

```python
df_model = df.groupby(['모델명','세부모델']).mean()
```

```python
df_model1 = df_model.drop(['Unnamed: 0','index','압류/저당','사고유무','전손이력'], axis=1)
```

```python
df_model2 = df_model1.drop(['배기량','연식','주행거리'], axis=1)
```

```python
df_model2
```

```python
df_model2.to_csv('./model.csv', encoding='cp949')
```
