```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time
import pandas as pd
pd.set_option('display.max_columns', None) # 데이터프레임 컬럼 display
from selenium.webdriver.common.keys import Keys
import warnings
warnings.filterwarnings('ignore') # 경고 무시
```

```python
def make_info2():
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    info=[]
    # name
    try:
        name_selector='#container > div.detail-wrap > div.tit-wrapper > ul > li.tit > h3'
        name = soup.select_one(name_selector).text.strip()
        info.append(name)
    except:
        info.append(" ")
        
    # 가격
    try:
        price_selector='#container > div.detail-wrap > div.tit-wrapper > ul > li.fr > p'
        price = soup.select_one(price_selector).text.strip()
        info.append(price)
    except:
        info.append(" ")
    
    #배기량
    try:
        smoke_selector='#scroll-tab1 > div > table > tbody > tr:nth-child(2) > td:nth-child(4)'
        smoke=soup.select_one(smoke_selector).text.strip()
        info.append(smoke)
    except:
        info.append(' ')
    
    # 연식
    try:
        age_selector='#scroll-tab1 > div > table > tbody > tr:nth-child(2) > td:nth-child(2)'
        age=soup.select_one(age_selector).text.strip()
        info.append(age)
    except:
        info.append(" ")
    
    #색상
    try:
        color_selector='#scroll-tab1 > div > table > tbody > tr:nth-child(1) > td:nth-child(8)'
        color=soup.select_one(color_selector).text.strip()
        info.append(color)
    except:
        info.append(' ')
    
    #주행거리
    try:
        distance_selector='#scroll-tab1 > div > table > tbody > tr:nth-child(3) > td:nth-child(2)'
        distance=soup.select_one(distance_selector).text.strip()
        info.append(distance)
    except:
        info.append(" ")
    
    #변속기
    try:   
        speedchanger_selector='#scroll-tab1 > div > table > tbody > tr:nth-child(1) > td:nth-child(6)'
        speedchanger=soup.select_one(speedchanger_selector).text.strip()
        info.append(speedchanger)
    except:
        info.append(" ")
    
    #연료
    try:
        oil_selector='#scroll-tab1 > div > table > tbody > tr:nth-child(1) > td:nth-child(4)'
        oil=soup.select_one(oil_selector).text.strip()
        info.append(oil)
    except:
        info.append(" ")
        
    #차종
    try:
        kind_selector='#scroll-tab1 > div > table > tbody > tr:nth-child(3) > td:nth-child(4)'
        kind=soup.select_one(kind_selector).text.strip()
        info.append(kind)
    except:
        info.append(' ')
        
    #사고유무
    try:
        acc_selector='#scroll-tab1 > div > table > tbody > tr:nth-child(2) > td:nth-child(6)'
        acc=soup.select_one(acc_selector).text.strip()
        info.append(acc)
    except:
        info.append(' ')

    #압류, 저당 -> 압류 정당이 한테이블
    try:
        seize_selector='#scroll-tab1 > div > table > tbody > tr:nth-child(2) > td:nth-child(8)'
        seize=soup.select_one(seize_selector).text.strip()
        info.append(seize)
    except:
        info.append(' ')
        
    #전손이력
    try:
        tloss_selector='#scroll-tab1 > div > div.state-wrap.fl > table > tbody > tr:nth-child(4) > td > table > tbody > tr > td:nth-child(1)'
        tloss=soup.select_one(tloss_selector).text.strip()
        info.append(tloss)
    except:
        info.append(' ')
        
    #침수전손
    try:
        water_selector='#scroll-tab1 > div > div.state-wrap.fl > table > tbody > tr:nth-child(4) > td > table > tbody > tr > td:nth-child(3)'
        water=soup.select_one(water_selector).text.strip()
        info.append(water)
    except:
        info.append(' ')
            
    #판매회사
    try:
        company_selector='#scroll-tab3 > div > table > tbody > tr:nth-child(1) > td:nth-child(2)'
        company=soup.select_one(company_selector).text.strip()
        info.append(company)
    except:
        info.append(' ')
        
    #판매지역
    try:
        location_selector='#scroll-tab3 > div > table > tbody > tr:nth-child(3) > td > a'
        location=soup.select_one(location_selector).text.strip()
        info.append(location)
    except:
        info.append(' ')
        
    #옵션
    #상세 옵션 버튼 클릭
    option = []
#     option_selector='#scroll-tab1 > div > div.option-cate > div > div > button'
#     cursor = driver.find_element_by_css_selector(option_selector)
#     cursor.click()
#     time.sleep(2)
    try:
        #option_list = soup.select("#scroll-tab1 > div > div.option-cate > div > div.cate-list-wrap.fade-enter-done > ul")
        #option_on=option_list.select_one(option_selector).text.strip()
        option_on = soup.select('.on span')
        for op in option_on:
            option.append(op.text)
        print(option)
        info.append(option)
    except:
        info.append(" ")

    return info
```

```python
# 중고차 url 접속
driver = webdriver.Chrome()
url = "https://autobell.co.kr/buycar/buyCarList"
driver.get(url)
time.sleep(2)

# 수입차 체크 해제
chbox = "#container > div.content-wrap.buy-wrap > div.search-sec > ul > li:nth-child(1) > dl > dd > p > span:nth-child(2) > label"
cursor = driver.find_element_by_css_selector(chbox)
cursor.click()
time.sleep(1)

df=[]
for i in range(1,2824):
    try:
        # 중고차 테이블 클릭
        selec = f'#tab8-2 > table > tbody > tr:nth-child({i}) > td:nth-child(2) > div > a > h5'
        cursor = driver.find_element_by_css_selector(selec)
        cursor.click()
        driver.switch_to.window(driver.window_handles[1])
        time.sleep(2)
        
        # 데이터 추출
        df.append(make_info2())
        time.sleep(3)
        
        # 창 닫기, 창 전환
        driver.close()
        driver.switch_to.window(driver.window_handles[0])
    except:
        # 중고차 테이블 없을시 더보기 버튼 클릭
        botton = "#tab8-2 > div > button"
        cursor = driver.find_element_by_css_selector(botton)
        cursor.click()
```
