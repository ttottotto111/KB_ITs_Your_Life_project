```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time
import pandas as pd
pd.set_option('display.max_columns', None) # 데이터프레임 컬럼 display
from selenium.webdriver.common.keys import Keys
import warnings
warnings.filterwarnings('ignore') # 경고 무시
```

```python
def make_info2():
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    info=[]
    # name
    try:
        name_selector='div.car-dt-info > div.car-buy-info > div.car-buy-price > strong.car-buy-name'
        name = soup.select_one(name_selector).text.strip()
        info.append(name)
    except:
        info.append(" ")
    # 차량번호
    try:
        number_selector='div.car-detail-info > div > div.detail-info01 > table > tbody > tr:nth-child(1) > td:nth-child(2)'
        number=soup.select_one(number_selector).text.strip()
        info.append(number)
    except:
        info.append(" ")
    # 연식
    try:
        age_selector='div.car-detail-info > div > div.detail-info01 > table > tbody > tr:nth-child(1) > td:nth-child(4)'
        age=soup.select_one(age_selector).text.strip()
        info.append(age)
    except:
        info.append(" ")
    #주행거리
    try:
        distance_selector='div.car-detail-info > div > div.detail-info01 > table > tbody > tr:nth-child(2) > td:nth-child(2)'
        distance=soup.select_one(distance_selector).text.strip()
        info.append(distance)
    except:
        info.append(" ")
    #연료
    try:
        oil_selector='div.car-detail-info > div > div.detail-info01 > table > tbody > tr:nth-child(2) > td:nth-child(4)'
        oil=soup.select_one(oil_selector).text.strip()
        info.append(oil)
    except:
        info.append(" ")
    #변속기
    try:   
        speedchanger_selector='div.car-detail-info > div > div.detail-info01 > table > tbody > tr:nth-child(3) > td:nth-child(2)'
        speedchanger=soup.select_one(speedchanger_selector).text.strip()
        info.append(speedchanger)
    except:
        info.append(" ")
    #연비
    try:
        fuele_selector='div.car-detail-info > div > div.detail-info01 > table > tbody > tr:nth-child(3) > td:nth-child(4)'
        fuele=soup.select_one(fuele_selector).text.strip()
        info.append(fuele)
    except:
        info.append(' ')
    #차종
    try:
        kind_selector='div.car-detail-info > div > div.detail-info01 > table > tbody > tr:nth-child(4) > td:nth-child(2)'
        kind=soup.select_one(kind_selector).text.strip()
        info.append(kind)
    except:
        info.append(' ')
    #배기량
    try:
        smoke_selector='div.car-detail-info > div > div.detail-info01 > table > tbody > tr:nth-child(4) > td:nth-child(4)'
        smoke=soup.select_one(smoke_selector).text.strip()
        info.append(smoke)
    except:
        info.append(' ')
    #색상
    try:
        color_selector='div.car-detail-info > div > div.detail-info01 > table > tbody > tr:nth-child(5) > td:nth-child(2)'
        color=soup.select_one(color_selector).text.strip()
        info.append(color)
    except:
        info.append(' ')
    #세금미납
    try:
        tax_selector='div.car-detail-info > div > div.detail-info01 > table > tbody > tr:nth-child(5) > td:nth-child(4)'
        tax=soup.select_one(tax_selector).text.strip()
        info.append(tax)
    except:
        info.append(' ')
    #압류
    try:
        seize_selector='div.car-detail-info > div > div.detail-info01 > table > tbody > tr:nth-child(6) > td:nth-child(2)'
        seize=soup.select_one(seize_selector).text.strip()
        info.append(seize)
    except:
        info.append(' ')
    #저당
    try:
        mortgage_selector='div.car-detail-info > div > div.detail-info01 > table > tbody > tr:nth-child(6) > td:nth-child(4)'
        mortgage=soup.select_one(mortgage_selector).text.strip()
        info.append(mortgage)
    except:
        info.append(' ')
    #전손이력
    try:
        tloss_selector='div.car-detail-info > div > div.detail-info02 > div > dl > dd:nth-child(2)'
        tloss=soup.select_one(tloss_selector).text.strip()
        info.append(tloss)
    except:
        info.append(' ')
    #침수이력
    try:
        water_selector='div.car-detail-info > div > div.detail-info02 > div > dl > dd:nth-child(4)'
        water=soup.select_one(water_selector).text.strip()
        info.append(water)
    except:
        info.append(' ')
    #용도이력
    try:
        use_selector='div.car-detail-info > div > div.detail-info02 > div > dl > dd:nth-child(6)'
        use=soup.select_one(use_selector).text.strip()
        info.append(use)
    except:
        info.append(' ')
    #소유자변경
    try:
        user_selector='div.car-detail-info > div > div.detail-info02 > div > dl > dd:nth-child(8)'
        user=soup.select_one(user_selector).text.strip()
        info.append(user)
    except:
        info.append(' ')
    #판매회사
    try:
        company_selector='div.seller-info > div > div.dealer-location > div.map-txt > p:nth-child(1)'
        company=soup.select_one(company_selector).text.strip()
        info.append(company)
    except:
        info.append(' ')
    #판매지역
    try:
        location_selector='div.seller-info > div > div.dealer-location > div.map-txt > p:nth-child(2)'
        location=soup.select_one(location_selector).text.strip()
        info.append(location)
    except:
        info.append(' ')
    #옵션
    option_selector='div.car-option > ul'
    try:
        option=soup.select_one(option_selector).text.strip()
        info.append(option)
    except:
        info.append(" ")
    #차체보증
    insure_selector='#divBanner3 > div.etc-sec01 > div > div:nth-child(1) > div'
    try:
        insure=soup.select_one(insure_selector).text.strip()
        info.append(insure)
    except:
        info.append(" ")
    #엔진보증
    engine_selector='#divBanner3 > div.etc-sec01 > div > div:nth-child(2) > div'
    try:
        engine=soup.select_one(engine_selector).text.strip()
        info.append(engine)
    except:
        info.append(" ")
    return info
```
